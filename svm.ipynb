```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df= pd.read_csv('data/data.csv')
df #dataset before preprocessing
```

```python
df.drop(['Unnamed: 32','id'],axis=1,inplace=True)# Dropped null column and id as its irrelevant to the project
```

```python
df['diagnosis']=df['diagnosis'].map({'M': 1, 'B': 0}) # converting categorical value of diagnosis to binary value
df
```

# Splitting Predictors and target

```python
X=df.drop(['diagnosis'],axis=1)
Y=df['diagnosis']
train_data=X.join(Y)
train_data
```

# Normalising data

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,Y,test_size=0.2,random_state=55)
```

# Tuning hyperparameters resulted in the following 
## Best Parameters:-
- 'C': 0.1,
- 'gamma': 0.01,
- 'kernel': 'linear'

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [0.01, 0.1, 1, 'scale'],
    'kernel': ['rbf','linear']
}

grid = GridSearchCV(SVC(), param_grid, cv=5, scoring='accuracy')
grid.fit(X_train, y_train)

print("Best Parameters:", grid.best_params_)
print("Best CV Score:", grid.best_score_)
```

# Training SVC

## Linear SVM with best parameters

```python
svc_lin=SVC(kernel='linear',gamma=0.01,C=0.1 )
svc_lin.fit(X_train,y_train)
y_pred=svc_lin.predict(X_test)
```

## rbf SVM with best parameters

```python
svc_rbf=SVC(kernel='rbf',gamma=0.01,C=0.1)
svc_rbf.fit(X_train,y_train)
y_pred1=svc_rbf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
# linear
accu_lin=accuracy_score(y_pred,y_test)
cr_lin=classification_report(y_pred,y_test)
# rbf
accu_rbf=accuracy_score(y_pred1,y_test)
cr_rbf=classification_report(y_pred1,y_test)

print('Accuracy score of linear model:',accu_lin)
print('Accuracy score of rbf model:',accu_rbf)

print("classification_report of linear model:-\n",cr_lin)
print("classification_report of rbf model:-\n",cr_rbf)
```

# Evaluation using Cross Validation scores

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(SVC(kernel='linear', C=.1, gamma=0.01),
                         X_train, y_train, cv=5, scoring='accuracy')

print("CV scores:", scores)
print("Mean CV accuracy:", scores.mean())
```

# Plotting Decision Boundary for most successful model

```python
X_2d = df[['perimeter_mean', 'area_mean']].values # picking two features at random
y=df['diagnosis'].values

scaler = StandardScaler()
X2dscaled = scaler.fit_transform(X_2d)
X_train, X_test, y_train, y_test = train_test_split(X2dscaled, y, test_size=0.2, random_state=29)

svc_lin = SVC(kernel='linear',gamma=0.01,C=0.1) #best parameters
svc_lin.fit(X_train, y_train)

x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                     np.linspace(y_min, y_max, 500))

Z = svc_lin.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=plt.cm.coolwarm, edgecolors='k')
plt.xlabel('perimeter_mean (scaled)')
plt.ylabel('area_mean (scaled)')
plt.title('SVM Decision Boundary (Linear Kernel)')
plt.show()
```
